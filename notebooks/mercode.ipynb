2. Leemos el csv y lo guardamos en un dataframe

```python
df = pd.read_csv('encuesta100.csv')
```

3. Limpieza de datos

```python
#vemos cuántos nulos hay en cada columna
print(df.isnull().sum())
```

4.Normalizamos los datos

```python
# Convertimos la fecha a formato datetime para que sea siempre igual
df = df.drop_duplicates() # eliminamos filas duplicadas
df.columns = df.columns.str.lower() # convertimos los nombres de las columnas a minúsculas
df.columns = df.columns.str.strip() # eliminamos espacios en blanco en los nombres de las columnas
df.fillna(0, inplace=True) # rellenamos valores nulos con 0
df.columns = df.columns.str.replace(' ', '_') # reemplazamos espacios en blanco por guiones bajos en los nombres de las columnas
df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce")
```

```python
print(df.isnull().sum())  # Verificamos que no haya nulos
print(df.info()) # mostramos informacion del dataframe luego de la limpieza
print(df.describe())# mostramos estadisticas descriptivas del dataframe
```
