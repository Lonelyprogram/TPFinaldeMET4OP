2. Leemos el csv y lo guardamos en un dataframe

```python
df = pd.read_csv('encuesta100.csv')
```

3. Limpieza de datos

```python
#vemos cuántos nulos hay en cada columna
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
df.columns = df.columns.str.lower() #ponemos todo en minusculas
df.columns = df.columns.str.strip() #eliminamos espacios al inicio y final de los nombres de las columnas
df.fillna(0) # rellenamos los valores nulos con 0 (en nuestro ejemplo no hay pero en otros casos puede ser util)
df = df.drop_duplicates() # eliminamos filas duplicadas
print(df.info()) # mostramos informacion del dataframe luego de la limpieza

# mostramos estadisticas descriptivas del dataframe
print(df.describe()) 
```

4.Normalizamos los datos

```python
# Convertimos la fecha a formato datetime para que sea siempre igual
df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce") 
# Convertimos numéricos
numericas = ["edad", "cantidad_de_integrantes_en_el_hogar", "imagen_del_candidato"]
for col in numericas:
    df[col] = pd.to_numeric(df[col], errors="coerce")
# Convertimos categóricos
categoricas = ["estrato", "sexo", "nivel_educativo", "voto", "voto_anterior"]
for col in categoricas:
    df[col] = df[col].astype("category")
```
